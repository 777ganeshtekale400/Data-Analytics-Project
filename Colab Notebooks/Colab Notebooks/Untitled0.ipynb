# **Step3 Check Dataset**

# **Step 4 Initialization**

```python
d=0.506
r=d/2
A=np.pi*r**2

stress_al=(df_al['FORCE']/A)*0.001
strain_al=df_al['CH5']*0.01

stress_steel=(df_steel['FORCE']/A)*0.001
strain_steel=df_steel['CH5']*0.01
```

# **Step 5 Plot Stress Strain Curve**

```python
fig,ax=plt.subplots()
ax.plot(strain_al,stress_al)
ax.plot(strain_steel,stress_steel)
ax.set_xlabel('Strain (mm/mm)')
ax.set_ylabel('Stress (kN/mm2)')
ax.set_title('Stress Strain Curve for AL 6061 & Steel 1018 in Tension')
ax.legend(['AL6061','Steel1018'])

plt.show()
```

# **Step 6 Calculate Tensile Strength**

```python
ts_al=np.max(stress_al)
ts_steel=np.max(stress_steel)
final_ts_steel=6.8976*ts_steel
print('The Tensile Strength of Steel 1018 in kN/mm2 is', final_ts_steel)
final_ts_steel=6.8976*ts_al
print('The Tensile Strength of Al 6061 in kN/mm2 is', final_ts_steel)
```
